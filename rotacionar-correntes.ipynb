<p> $\omega$ -> Direção da corrente medida pelo ADCP em relação ao norte geográfico(sentido horário).</p>
<p>
$\phi$ -> Direção da corrente em relação ao Leste (angulo que devemos usar para computar u e v)
</p>
<p>$\theta$ ->  É o angulo formado entre os eixos x' e x (Leste) que é a inclinação das linhas de batimetria ou do canal principal  em relação ao L.</p>

### Cálculo de u e v

<p>$\phi = 90-\omega$</p>
<p> $ u = r\cdot cos(\phi)$</p>
<p>$ v = r\cdot sin(\phi)$</p>

## rotação no sentido anti-horário
<p>$\theta > 0$</p>

<p> $u' = u\cdot cos(\theta) + v\cdot sin(\theta)$ </p>
<p> $v' = v\cdot cos(\theta) - u \cdot sin(\theta)$ </p>

## rotação no sentido horário
<p> $\theta < 0$</p>

<p>$v' = v\cdot \cos(\theta) + u\cdot sin(\theta)$</p>
<p>$u' = u\cdot \cos(\theta) - v\cdot sin(\theta)$</p>

## Função

```python
def rotaciona_corrente(u, v, theta):
    """
    Retorna as componentes u' e v' de uma corrente
    no plano x', y' rotacionado theta(graus) em 
    relação ao Leste/Norte a partir das componentes
    u(zonal) e v(meridional). Ref: Emery e Thomson
    2014.
    """

    if theta > 0:
        theta = np.deg2rad(theta)
        ul = u * np.cos(theta) + v * np.sin(theta)
        vl = v * np.cos(theta) - u * np.sin(theta)
        
    
    elif theta < 0:
        theta = np.deg2rad(theta)
        ul = u * np.cos(theta) - v * np.sin(theta)
        vl = v * np.cos(theta) + u * np.sin(theta)
        
    return ul, vl
```
